## 一、概览
　　Apktool用于用于解析apk的res文件、smali文件以及AndroidManifest.xml文件

## 二、安装Apktools
　　确保在安装 Apktools 之前，已经安装好 Java JDK，并配置好Java开发环境。  
### 2.1、下载安装：
1、下载启动文件（[wrapper script](https://raw.githubusercontent.com/iBotPeaches/Apktool/master/scripts/windows/apktool.bat)）,并另存为apktool.bat。  
2、下载最新版本的[apktool_xxx.jar](https://bitbucket.org/iBotPeaches/apktool/downloads/),并重命名为apktool.jar。  
3、将下载好的apktool.bat、apktool.jar移动到同一目录下。  
4、配置环境变量：添加Apktool文件的目录到Path下。  
5、打开 cmd 命令行。输入apktool回车，验证安装。

## 三、apktool使用
### 3.1、反编译apk文件

'''shell
java -jar apktool.jar d -f Example.apk -o Example   # 未配置apktool环境变量  
apktool.jar d -f Example.apk -o Example  # 配置了apktool环境变量
apktool d Example.apk
'''
反编译完成后，得到的res文件夹是程序当中的所有资源文件，smali文件夹则存放了程序所有的反汇编代码。
### 3.2、重新打包apk
'''shell
apktool.bat b Example\ -o Example.apk    
apktool b Example\ Example.apk  
'''

### 3.3、重新签名
　　在新生成的 Example.apk 目录下，将 apktool.jar 复制到该目录，然后重新生成签名文件。  
'''shell
keytool -genkey -alias example.keystore -keyalg RSA -validity 20000 -keystore example.keystore
'''
回车后，前两个是要你输入密码，后面一些名称之类的随便填，最后输入 y 回车，生成 keystore。  
使用keystore对apk进行签名：

'''shell
jarsigner -verbose -keystore example.keystore -signedjar Example_1.0.apk Example.apk example.keystore  
'''
后面的两个apk，前一个是 要生成的apk，后一个是1.3步骤打包生成的apk。
