
## 二、Android 逆向之反Root检测
　　Zygisk版本面具Magisk虽然移除了MagiskHide但也可以实现隐藏Root。
[新版Magisk(面具)隐藏root方法](https://zhuanlan.zhihu.com/p/506203078)   
[隐藏Root - Zygisk版面具Magisk过银行App等Root检测](https://www.bilibili.com/read/cv15350941/)   

### 2.1、方法一

1、打开magisk，点击右上角设置，找到Zygisk选项并且打开，打开后重启手机。
2、重启手机后刷入Shamiko模块并开启，重启手机，打开面具右上角设置，打开‘遵循排除列表’选项，再点击‘配置排除列表’，勾选你要隐藏root的应用（注意应用下的所有选项都要勾选），然后关闭‘遵守排除列表’选项，重启手机。（如果没有关闭‘遵循排除列表’，此时，虽然可以绕过root检测，但是在排除列表中的应用是无法使用虚拟框架和模块的）     
　　（去浏览器中下载Shamiko模块并保存到方便的位置，在面具Magisk的模块功能中使用本地安装刷入该模块）   
3、一般来说，完成上面的操作步骤就可以了，如果发现有app还是会提示设备已root，那么就要随机Magisk包名以达到隐藏Magisk的效果，以防止被某些软件检测到Magisk判别root。
4、隐藏Magisk后，此时大部分的软件都已经检测不到root，但是还有一部分顽强的软件还是可以检测到root十分顽强，那么我们该怎么办呢？我们下载Zygisk版本的LSPosed，用面具刷入，重启手机。下载隐藏应用列表，然后打开隐藏应用列表，点击检测测试下载测试软件，后面需要用到和点击安装magisk插件，选择Zygisk，安装成功后，重启手机。 然后点击模板管理，创建黑名单模板，填写以下可能被检测到的maps规则

如下图将规则复制粘贴：
```txt
magisk
root
lsposed
edxposed
xposed
qn_dyn_lib
qn_mmkv
org.meowcatedxposedmanager
org.lsposed.manage
manager
com.topjohnwumagisk
magisk manager
```
　　然后然后点击隐藏应用，勾选带有root权限的所有软件。保存刚才设置好的规则,选择你要隐藏root的软件，按照下面的图片并且保存选择刚才设置好的规则模板，保存即可。
　　还要注意的是，那个工作模式不要盲目选择所有隐藏方式，容易让某些软件直接崩溃 下面打开测试软件，我们会发现，没有被检测到root，这样说明很多软件都不会被检测了，可以正常使用。
　　magisk的延伸版本 Canary（金丝雀版本）和alpha（阿尔法版本）隐藏方法同理，还要注意一点Zygisk不能和riru系列的模块共存，安装riru，Shamiko模块就会不工作，如果你想用riru系列模块又想隐藏root建议会退magisk23.0即可。
