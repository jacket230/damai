## 一、概述
　　crypto-js(GitHub)是谷歌开发的一个纯JavaScript的加密算法类库，可以非常方便的在前端进行其所支持的加解密操作。目前crypto-js已支持的算法有：MD5,SHA-1,SHA-256,AES,Rabbit,MARC4,HMAC,HMAC-MD5,HMAC-SHA1,HMAC-SHA256,PBKDF2。常用的加密方式有MD5和AES，使用时可以引用总文件，也可以单独引用某一文件。<br>
　　**安装使用：**<br>
　　　　直接引用：引入crypto-js文件。<br>
　　　　安装依赖：npm install crypto-js　　# 使用npm安装,安装之后看到node_modules目录直接放到public目录下<br>
　　　　　　　　　bower install crypto-js　　# 使用Bower安装<br>

## 二、引用方式
　　CryptoJS的使用分为两种环境：Node.js服务器端和浏览器端。<br>
### 2.1、在Node.js里使用
#### 2.1.1、ES6方式
>import CryptoJS from 'crypto-js';<br>
import sha256 from 'crypto-js/sha256';<br>
import hmacSHA512 from 'crypto-js/hmac-sha512';<br>
import Base64 from 'crypto-js/enc-base64';<br>
const message, nonce, path, privateKey; // ...<br>
const hashDigest = sha256(nonce + message);<br>
const hmacDigest = Base64.stringify(hmacSHA512(path + hashDigest, privateKey));<br>

#### 2.1.2、非ES6方式
>var AES = require("crypto-js/aes");<br>
var SHA256 = require("crypto-js/sha256");<br>
...<br>
console.log(SHA256("Message"));<br>

可以引入整个CryptoJS
>var CryptoJS = require("crypto-js");<br>
console.log(CryptoJS.HmacSHA1("Message", "Key"));<br>

### 2.2、在浏览器客户端使用
#### 2.2.1、RequireJS方式
　　**导入CryptoJS模块**
>require.config({<br>
　　packages: \[<br>
　　　　{<br>
　　　　　　name: 'crypto-js',<br>
　　　　　　location: 'path-to/bower_components/crypto-js',<br>
　　　　　　main: 'index'<br>
　　　　}<br>
　　\]<br>
});<br>
require(\["crypto-js/aes", "crypto-js/sha256"\], function (AES, SHA256) {<br>
　　console.log(SHA256("Message"));<br>
});<br>

　　导入整个CryptoJS模块，这样可以使用所有加解密的方法
>// Above-mentioned will work or use this simple form<br>
require.config({<br>
　　paths: {<br>
　　　　'crypto-js': 'path-to/bower_components/crypto-js/crypto-js'<br>
　　}<br>
});<br>
require(\["crypto-js"\], function (CryptoJS) {<br>
　　console.log(CryptoJS.HmacSHA1("Message", "Key"));<br>
});<br>

#### 2.2.2、非RequireJS方式
>\<script type="text/javascript" src="path-to/bower_components/crypto-js/crypto-js.js"\>\</script\><br>
\<script type="text/javascript"\><br>
    var encrypted = CryptoJS.AES(...);<br>
    var encrypted = CryptoJS.SHA256(...);<br>
\</script\><br>

## 三、CryptoJS API
　　这里以AES加密，加密的对象分为两种情况：纯文本和对象。
### 3.1、纯文本加密
>var CryptoJS = require("crypto-js");<br>
// Encrypt<br>
var ciphertext = CryptoJS.AES.encrypt('my message', 'secret key 123').toString();<br>
// Decrypt<br>
var bytes  = CryptoJS.AES.decrypt(ciphertext, 'secret key 123');<br>
var originalText = bytes.toString(CryptoJS.enc.Utf8);<br>
console.log(originalText); // 'my message'<br>
### 3.2、对象加密
>var CryptoJS = require("crypto-js");<br>
var data = \[{id: 1}, {id: 2}\]<br>
// Encrypt<br>
var ciphertext = CryptoJS.AES.encrypt(JSON.stringify(data), 'secret key 123').toString();<br>
// Decrypt<br>
var bytes  = CryptoJS.AES.decrypt(ciphertext, 'secret key 123');<br>
var decryptedData = JSON.parse(bytes.toString(CryptoJS.enc.Utf8));<br>
console.log(decryptedData); // \[{id: 1}, {id: 2}\]<br>
